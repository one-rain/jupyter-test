```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import decimal
import sys
import os


def myprint(msg):
    print("  ")
    print("============== %s ==============" % msg)
    print("  ")


def demo_duplicated():
    myprint("去掉重复行")
    df = pd.read_excel(r'D:\data\sport\iqiyi_pingback.xls', 'Sheet1', keep_default_na=False)

    df['plat'] = df['plat'].map(lambda e: str.lower(e))
    df['page_code'] = df['page_code'].map(lambda e: e.strip('_').strip('.'))
    df['block_code'] = df['block_code'].map(lambda e: str(e).strip('_').strip('.'))

    df1 = df[['plat', 'os', 'page_name', 'page_code']]
    print(df1.drop_duplicates())


def get_data():
    company = ['A', 'B', 'C']
    data = {"company": [company[x] for x in np.random.randint(0, len(company), 10)],
            "salary": np.random.randint(5, 50, 10),
            "age": np.random.randint(15, 50, 10)
            }
    return pd.DataFrame(data)


def demo_group1():
    myprint("分组1")
    df = get_data()

    # groupby就是将原有的 DataFrame按照groupby的字段，划分为若干个分组DataFrame
    # 被分为多少个组就有多少个分组DataFrame
    print(list(df.groupby('company')))

    myprint('agg - all mean')
    # 按照company求salary和age的平均值
    print(df.groupby('company').agg('mean'))

    myprint('agg - salary:median, age: mean')
    # 按照company求salary的最大值和age的平均值
    print(df.groupby('company').agg({'salary':'median', 'age':'mean'}))
    
  
def demo_group2():
    # 在原数据基础上增加一列，代表员工所在的公司的平均薪水
    myprint("分组2")
    df = get_data()

    df = get_data()
    avg_salary_dict = df.groupby('company')['salary'].mean().to_dict()
    # print(avg_salary_dict)
    df['avg_salary'] = df['company'].map(avg_salary_dict)

    print(list(df.groupby('company')))

    myprint('使用transform处理')
    # transform会在分完组内数据执行指定的函数，然后按照原索引的顺序返回结果
    df['avg_salary'] = df.groupby('company')['salary'].transform('mean')
    print(list(df.groupby('company')))


def demo_group3():
    # 获取各个公司年龄最大的员工的数据
    myprint("分组3")
    df = get_data()
    print(list(df.groupby('company')))

    myprint('各公司年龄最大的')
    oldest_staff = df.groupby('company', as_index=False).apply(lambda x: x.sort_values(by = 'age', ascending=True).iloc[-1,:])
    print(oldest_staff)


demo_group3()
```
