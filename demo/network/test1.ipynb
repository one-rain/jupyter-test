NetworkX生成有向无环图，并进行拓扑排序



```python
import networkx as nx
import matplotlib.pyplot as plt

print("有向图")
G = nx.DiGraph()
#G.add_nodes_from(['tableA', 'tableB', 'tableC', 'tableD', 'tableE', 'tableF', 'tableG', 'tableH', 'tableI'])
G.add_node('table0')
G.add_edges_from([("tableA", "tableB"), ("tableC", "tableB"), ("tableB", "tableD"), ("tableB", "tableE"), ("tableE", "tableF"), ("tableE", "tableG")])
G.add_edges_from([("tableH", "tableI"), ("tableI", "tableG"), ("tableG", "tableJ")])
G.add_edges_from([("tableM", "tableN"), ("tableN", "tableL"), ("tableN", "tableK")])
G.add_node('tableK')

nx.draw(G, with_labels=True)
plt.show()

print("")
print("==========拓扑排序==========")
print((list(nx.topological_sort(G))))

print("")
print("==========有子节点的节点==========")
for node in G.nodes():
    print("{}的子节点个数：{}，边缘个数：{}".format(node, G.out_degree(node), G.out_degree(node)))

```

```python
import networkx as nx
import matplotlib.pyplot as plt

print("无向图")
G = nx.Graph()
#G.add_nodes_from(['tableA', 'tableB', 'tableC', 'tableD', 'tableE', 'tableF', 'tableG', 'tableH', 'tableI'])
G.add_node('table0')
G.add_edges_from([("tableA", "tableB"), ("tableC", "tableB"), ("tableB", "tableD"), ("tableB", "tableE"), ("tableE", "tableF"), ("tableE", "tableG")])
G.add_edges_from([("tableH", "tableI"), ("tableI", "tableG"), ("tableG", "tableJ")])
G.add_edges_from([("tableM", "tableN"), ("tableN", "tableL"), ("tableN", "tableK")])
G.add_node('tableK')

color =['y', 'g', 'r']
subplot = [222, 223, 224]

plot = 221
plt.subplot(plot)
nx.draw_networkx(G, with_labels=True)

print("")
print("不连通分量：{}".format(nx.number_connected_components(G)))
print("==========不连通的子图==========")
for c in nx.connected_components(G):
    # nodeSet = G.subgraph(c).nodes()
    # plot = plot + 1
    print(c)
    plt.subplot(subplot[0])
    nx.draw_networkx(G.subgraph(c), with_labels=True, node_color=color[0])
    color.pop(0)
    subplot.pop(0)
    
plt.show()

```
